## 2. Get specific contents using BeatifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

### 2.2 Get the first paragraph

```python
import numpy as np
import pandas as pd
data1 = []
table = soup.find('table')
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    #print(cols)
    for ele in cols:
        if ele.get_text():
            data1.append(ele.get_text(separator = ', '))
data1
```

```python
df = pd.DataFrame(np.reshape(data1[5:], (-1, 4))) 
df.columns = ['ranking', 'beer_info', 'rating#', 'ave_score']

df
```

```python
# beer_info column contains many different inforamtion separated by ','. I split the columns and add to df.
# the original column is still here
# there is one additional column contains all None value, since there is ',' at the end in beer_info and inevitably used for split.
df[['beer_name', 'company', 'type', 'alcohol']] = df['beer_info'].str.split(', ', n=3, expand=True)
df
```
